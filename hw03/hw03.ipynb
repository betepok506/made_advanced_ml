# 1. Базовый частотный метод по Шерлоку Холмсу

Реализуйте базовый частотный метод по Шерлоку Холмсу:
   * подсчитайте частоты букв по корпусам (пунктуацию и капитализацию можно просто опустить, а вот пробелы лучше оставить);
   * возьмите какие-нибудь тестовые тексты (нужно взять по меньшей мере 2-3 предложения, иначе вряд ли сработает), зашифруйте их посредством случайной перестановки символов; 
   * расшифруйте их таким частотным методом.

```python
sns.set_style("whitegrid")
palette = sns.color_palette()
figsize = (15, 8)
legend_fontsize = 18
axis_fontsize = 16

random.seed(42)

PATH_TO_DATA = "./WarAndPeace.txt"
PATH_TO_TEST_TEXT = './AnnaKarenina.txt'

a = ord('а')
ALPHABET = ''.join([chr(i) for i in range(a,a+32)])
ALPHABET += ' '
```

```python
ALPHABET
```

```python
with open(PATH_TO_DATA, 'r', encoding="utf-8") as file:
    corpus_rus_text = ' '.join(file.readlines())
```

```python
def preprocess_text(text, exp):
    text = text.lower()
    text = re.sub(exp, ' ', text)
    text = re.sub(' +', ' ', text)
    return text
```

```python
'''
    Предобработаем текст, приведя его в нижний регистр 
    и оставив только русские буквы
'''
print(f"Количество символов до предобработки текста: {len(corpus_rus_text)}")
corpus_rus_text = preprocess_text(corpus_rus_text, "[^а-я ]")
print(f"Количество символов после предобработки текста: {len(corpus_rus_text)}")
```

```python
# Словарь частот символов романа "Война и мир"
frequency_occurrence_letters = dict(Counter(corpus_rus_text))
```

```python
frequencies_list = [frequency_occurrence_letters[letter] for letter in ALPHABET]

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=figsize)

ax.bar(list(ALPHABET), frequencies_list)
ax.set_title(r'Частоты символов в романе "Война и мир"', fontsize=legend_fontsize)
ax.set_xlabel(r'Символы', fontsize=axis_fontsize)
ax.set_ylabel(r'Частота символов', fontsize=axis_fontsize)

plt.show()
```

```python
alphabet_encoding = list(ALPHABET)
random.shuffle(alphabet_encoding)
alphabet_encoding = ''.join(alphabet_encoding)
print(f"Оригинальный алфавит: {ALPHABET}")
print(f"Закодированный алфавит: {alphabet_encoding}")
```

```python
table_translation = str.maketrans(ALPHABET, alphabet_encoding)
```

```python
def get_alphabet(frequency_occurrence_letters, fl_sort=False):
    frequency_occurrence_letters = list(frequency_occurrence_letters.items())
    if fl_sort:
        frequency_occurrence_letters = sorted(frequency_occurrence_letters, key=lambda x: x[1], reverse=True)
    
    alphabet = "".join([x[0] for x in frequency_occurrence_letters])
    return alphabet


def encode_text(text, table_translation):
    '''Кодирование текста на основе перестановки оригинального алфавита'''
    return text.translate(table_translation)


def decode_text(text, frequency_occurrence_letters):
    '''Декодирование текста с помощью частотного метода по Шерлоку Холмсу'''
    original_alphabet = get_alphabet(frequency_occurrence_letters, True)
    encoding_frequencies = dict(Counter(text))
    for letter in original_alphabet:
        if letter not in encoding_frequencies:
            encoding_frequencies[letter] = 0
        encoding_frequencies[letter] += 1
            
    encoding_alphabet = get_alphabet(encoding_frequencies, True)
    table_translation = str.maketrans(encoding_alphabet, original_alphabet)
    return text.translate(table_translation)


def get_accuracy_decode(original_text, decode_text):
    '''Оценка точности декодирования текста'''
    num_matches = 0
    for x1, x2 in zip(original_text, decode_text):
        if x1 == x2:
            num_matches += 1
    return num_matches / len(original_text)
```

```python
with open(PATH_TO_TEST_TEXT, 'r', encoding='utf-8') as file:
    test_corpus = ' '.join(file.readlines())

test_corpus = preprocess_text(test_corpus, '[^а-я ]')
text_size = len(test_corpus)
```

```python
indexes = [31354, 713557, 582738, 326064]
for ind, simple_size in enumerate([3000, 10000, 40000, 400000]):
#     start_ind = random.randint(0, text_size - simple_size)
    start_ind = indexes[0]
    print(f"Индекс начала отрывка: {start_ind}")
    simple = test_corpus[start_ind: start_ind + simple_size]
    print(f"Размер отрывка: {simple_size}", end='\n\n')
    print(f"Оригинальный текст: {simple[:min(len(simple), 200)]}", end='\n\n')
    encoded_text = encode_text(simple, table_translation)
    print(f"Зашифрованный текст: {encoded_text[:min(len(encoded_text), 200)]}", end='\n\n')
    decoded_text = decode_text(encoded_text, frequency_occurrence_letters)
    print(f"Дешифрованный текст: {decoded_text[:min(len(decoded_text), 200)]}", end='\n\n')
    print(f"Точность декодирования текста: {get_accuracy_decode(simple, decoded_text)}", end='\n\n')
    print('-' * 60)
```

Из вывода сверху видно, что при размере отрывка длиной 40000 символов удалось добиться лучшего декодирования. Текст хорошо читаем и слова разлечимы. Однако при повышении размера отрывка до 400000 символов произошло ухудшение перевода

# 2 Частотный подход на основе биграмм

Вряд ли в результате получилась такая уж хорошая расшифровка, разве что если вы брали в качестве тестовых данных целые рассказы. Но и Шерлок Холмс был не так уж прост: после буквы E, которая действительно выделяется частотой, дальше он анализировал уже конкретные слова и пытался угадать, какими они могли бы быть. Я не знаю, как запрограммировать такой интуитивный анализ, так что давайте просто сделаем следующий логический шаг:

   * подсчитайте частоты биграмм (т.е. пар последовательных букв) по корпусам;
   * проведите тестирование аналогично п.1, но при помощи биграмм.

```python
def get_frequencies_bigram(text):
    _frequencies_bigram = Counter([text[i:i+2] for i in range(len(text) - 1)])
    for f_letter in ALPHABET:
        for s_letter in ALPHABET:
            _frequencies_bigram[f_letter + s_letter] += 1
            
    return dict(_frequencies_bigram)


def bigram_decode_text(text, frequencies_bigram):
    sorted_original_frequencies_bigram = sorted(list(frequencies_bigram.items()), key=lambda x: x[1], reverse=True)
    original_frequencies_bigram = [x[0] for x in sorted_original_frequencies_bigram]
    
    encoding_frequencies_bigram = get_frequencies_bigram(text)
    sorted_encoding_frequencies_bigram = sorted(list(encoding_frequencies_bigram.items()), key=lambda x: x[1], reverse=True)
    encoding_frequencies_bigram = [x[0] for x in sorted_encoding_frequencies_bigram]
    
    encoding2original = {}
    for f_bigram, s_bigram in zip(original_frequencies_bigram, encoding_frequencies_bigram):
        encoding2original[s_bigram] = f_bigram
    
    translation = []
    for i in range(0, len(text) - 1, 2):
        translation.append(encoding2original[text[i:i+2]])
        
    return "".join(translation)
```

```python
# Словарь частот биграмм романа "Война и мир"
frequency_occurrence_bigrams = get_frequencies_bigram(corpus_rus_text)
```

# 3 Частотный подход на основе биграмм (Улучшенная версия)

Но и это ещё не всё: биграммы скорее всего тоже далеко не всегда работают. Основная часть задания — в том, как можно их улучшить:
   * предложите метод обучения перестановки символов в этом задании, основанный на MCMC-сэмплировании, но по-прежнему работающий на основе статистики биграмм;
   * реализуйте и протестируйте его, убедитесь, что результаты улучшились.

```python
def get_table_translation(text, frequency_occurrence_letters):
    '''Функция для получения таблицы перевода одного алфавита в другой'''
    original_alphabet = get_alphabet(frequency_occurrence_letters, True)
    encoding_frequencies = dict(Counter(text))

    for letter in original_alphabet:
        if letter not in encoding_frequencies:
            encoding_frequencies[letter] = 0
        encoding_frequencies[letter] += 1
            
    encoding_alphabet = get_alphabet(encoding_frequencies, True)
    table_translation = dict(zip(encoding_alphabet, original_alphabet))
    return table_translation


def table_translation_shuffle(table_translation):
    new_table_translation = copy.copy(table_translation)
    symbol1, symbol2 = np.random.choice(list(new_table_translation.keys()), size=2, replace=False)
    new_table_translation[symbol1], new_table_translation[symbol2] = new_table_translation[symbol2], new_table_translation[symbol1]
    
    return new_table_translation


def log_likelihood(text, dict_translation, frequencies_bigram_dict, n_gram=2):
    """Подсчет вероятности строки на основе частот биграмм."""
    decode_text = translate_text(text, dict_translation)
    log_likelihood = 0
    for i in range(len(decode_text) - n_gram):
        log_likelihood += np.log(frequencies_bigram_dict[decode_text[i:i+n_gram]])
        
    return log_likelihood


def metropolis_hastings_log_accept(l, l_new):
    """Принимаем ли изменение в соответсвии на основе нового логарифма правдоподобия текста."""
    if l_new > l:
        return True
    else:
        return (np.random.rand() < (np.exp(l_new-l)))


def translate_text(text: str, table_translation):
    """Дешифровывает тест на основе таблицы-соответствия алфавитов."""
    return "".join([table_translation[x] for x in text]) 


def mcmc_decode_text(text, iterations, frequency_occurrence_letters, frequency_occurrence_bigrams, n_gram=2):
    """Метод обучения соответсвия между алфавитами, основанный на MCMC-сэмплировании."""

    curr_table_translation = get_table_translation(text, frequency_occurrence_letters)
    curr_l = log_likelihood(text, curr_table_translation, frequency_occurrence_bigrams, n_gram)
    for i in range(iterations):
        new_table_translation = table_translation_shuffle(curr_table_translation)
        new_l = log_likelihood(text, new_table_translation, frequency_occurrence_bigrams, n_gram)
        
        if metropolis_hastings_log_accept(curr_l, new_l):
            curr_table_translation, curr_l = new_table_translation, new_l
            
    return translate_text(text, curr_table_translation), curr_l
```

```python
indexes = [31354, 713557, 582738, 326064]
for ind, simple_size in enumerate([3000, 10000, 40000, 400000]):
#     start_ind = random.randint(0, text_size - simple_size)
    start_ind = indexes[0]
    print(f"Индекс начала отрывка: {start_ind}")
    simple = test_corpus[start_ind: start_ind + simple_size]
    print(f"Размер отрывка: {simple_size}", end='\n\n')
    print(f"Оригинальный текст: {simple[:min(len(simple), 200)]}", end='\n\n')
    encoded_text = encode_text(simple, table_translation)
    print(f"Зашифрованный текст: {encoded_text[:min(len(encoded_text), 200)]}", end='\n\n')
    decoded_text, _ = mcmc_decode_text(encoded_text, 3000, frequency_occurrence_letters, frequency_occurrence_bigrams)
    print(f"Дешифрованный текст: {decoded_text[:min(len(decoded_text), 200)]}", end='\n\n')
    print(f"Точность декодирования текста: {get_accuracy_decode(simple, decoded_text)}", end='\n\n')
    print('-' * 60)
```

Как можно видеть из результатов метод на основе семплирования МСМС позволил повысить точность до 100%.

```python
def get_frequencies_trigrams(text):
    _frequencies_trigrams = Counter([text[i:i+3] for i in range(len(text) - 2)])
    for f_letter in ALPHABET:
        for s_letter in ALPHABET:
            for th_letter in ALPHABET:
                _frequencies_trigrams[f_letter + s_letter + th_letter] += 1
            
    return dict(_frequencies_trigrams)
```
